```python
import pandas as pd
import os
```

```python
folder_path = "/content/drive/MyDrive/Tugas Akhir/Dataset/(1.4-Umum) Sampling Kalimat/SegmentedDataset"
output_path_full = os.path.join(folder_path, "Movie_Sampled_pipeline.csv")
output_path_only = os.path.join(folder_path, "Movie_Review_pipeline_smapled.csv")
```

```python
sample_percent = 0.00746
all_samples = []
```

```python
# List untuk log hasil sampling
sampling_log = []
```

```python
for filename in os.listdir(folder_path):
    if filename.endswith("_segmented.csv"):
        file_path = os.path.join(folder_path, filename)
        df = pd.read_csv(file_path)

        total_rows = len(df)
        n_samples = max(1, int(total_rows * sample_percent))
        sampled_df = df.sample(n=n_samples, random_state=42)
        sampled_df["Source_File"] = filename

        # Simpan log
        sampling_log.append({
            "File": filename,
            "Total_Review": total_rows,
            "Sampled": n_samples
        })

        all_samples.append(sampled_df)
```

```python
# Gabungkan semua hasil
combined_sample = pd.concat(all_samples, ignore_index=True)
```

```python
# Simpan file
combined_sample.to_csv(output_path_full, index=False)
```

```python
# Versi hanya Sentence
if "Sentence" in combined_sample.columns:
    combined_sample[["Sentence"]].to_csv(output_path_only, index=False)
else:
    print("Kolom 'Sentence' tidak ditemukan!")
```

```python
# Tampilkan ringkasan jumlah sampling
log_df = pd.DataFrame(sampling_log)
print("\n📊 Ringkasan jumlah data yang disample per file:")
print(log_df.to_string(index=False))
```
