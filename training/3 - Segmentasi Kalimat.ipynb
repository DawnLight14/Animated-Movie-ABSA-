```python
# ======== 2️⃣ Install dan Import Library ========

!pip install nltk pandas

import pandas as pd
import nltk
from nltk.tokenize import sent_tokenize
import os
import glob

# Download tokenizer untuk bahasa Inggris
nltk.download('punkt_tab')
```

```python
# ======== 3️⃣ Fungsi Segmentasi ========

def segment_sentences_in_folder(folder_path, output_folder):
    # Cari semua *_preprocessed.csv dalam folder
    csv_files = glob.glob(os.path.join(folder_path, '*_processed_gen.csv'))

    print(f'\n📂 Ditemukan {len(csv_files)} file CSV di folder: {folder_path}\n')

    # Buat folder output kalau belum ada
    os.makedirs(output_folder, exist_ok=True)

    total_sentences = 0

    for csv_file in csv_files:
        print(f'➡️ Processing file: {os.path.basename(csv_file)}')

        # Baca CSV
        df = pd.read_csv(csv_file)

        # Cek kolom Review
        if 'Review' not in df.columns:
            print(f'❌ Kolom "Review" tidak ditemukan di file {os.path.basename(csv_file)} — dilewati.\n')
            continue

        # Simpan hasil segmentasi
        segmented_data = []

        for idx, row in df.iterrows():
            movie_id = row['Movie_ID']
            movie_name = row['Movie_Name']
            date = row['Date']
            rating = row['Rating']
            review = row['Review']

            # Segmentasi paragraf → kalimat
            sentences = sent_tokenize(review)

            for sent in sentences:
                segmented_data.append({
                    'Movie_ID': movie_id,
                    'Movie_Name': movie_name,
                    'Date': date,
                    'Rating': rating,
                    'Sentence': sent.strip()
                })

        # Buat DataFrame hasil
        df_segmented = pd.DataFrame(segmented_data)

        # Nama output: [Movie_ID]_segmented.csv
        movie_id = str(df['Movie_ID'].iloc[0])
        save_path = os.path.join(output_folder, f'{movie_id}_segmented.csv')

        df_segmented.to_csv(save_path, index=False)

        print(f'✅ Simpan ke: {save_path}')
        print(f'➡️ Jumlah kalimat di file ini: {len(df_segmented)}\n')

        total_sentences += len(df_segmented)

    print('\n🚀 Proses selesai.')
    print(f'🔢 Total seluruh kalimat (semua file): {total_sentences}\n')
```

```python
# ======== 4️⃣ Jalankan ========

# Ganti path berikut sesuai lokasi folder kamu di Google Drive
input_folder = '/content/drive/MyDrive/Tugas Akhir/Dataset/V3/Disney/'
output_folder = '/content/drive/MyDrive/Tugas Akhir/Dataset/V4/Disney/'

# Jalankan segmentasi
segment_sentences_in_folder(input_folder, output_folder)
```
