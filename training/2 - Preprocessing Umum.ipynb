```python
# 2️⃣ Import library
import pandas as pd
import re
import os
```

```python
# 3️⃣ Fungsi preprocessing umum
def preprocessing_umum(text):
    # Hilangkan URL
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)

    # Hilangkan newline
    text = text.replace('\n', ' ').replace('\r', ' ')

    # Hapus spasi sebelum tanda baca
    text = re.sub(r'\s+([.,!?;:])', r'\1', text)

    # Tambahkan spasi setelah tanda baca jika belum ada (kecuali di akhir kalimat)
    text = re.sub(r'([.,!?;:])([^\s])', r'\1 \2', text)

    # Hilangkan spasi berlebih (ganda)
    text = re.sub(r'\s+', ' ', text)

    # Strip spasi di awal dan akhir
    text = text.strip()

    return text
```

```python
# 4️⃣ Fungsi proses folder CSV
def proses_folder_csv(folder_path, output_folder):
    # Buat folder output jika belum ada
    os.makedirs(output_folder, exist_ok=True)

    # Loop setiap file CSV di folder
    for filename in os.listdir(folder_path):
        if filename.endswith('_reviews_v2.csv'):
            file_path = os.path.join(folder_path, filename)
            print(f'Processing: {filename}')

            # Baca CSV
            df = pd.read_csv(file_path)

            # Cek apakah ada kolom 'Review'
            if 'Review' in df.columns:
                # Lakukan preprocessing umum → timpa kolom Review
                df['Review'] = df['Review'].astype(str).apply(preprocessing_umum)

                # Ambil Movie_ID dari filename
                movie_id = filename.split('_reviews_v2.csv')[0]
                output_filename = f'{movie_id}_processed_gen.csv'
                output_path = os.path.join(output_folder, output_filename)

                # Simpan ke CSV baru
                df.to_csv(output_path, index=False)
                print(f'Saved: {output_path}')
            else:
                print(f'Kolom "Review" tidak ditemukan di {filename}, dilewati.')
```

```python
# 5️⃣ Contoh pemanggilan fungsi
# GANTI PATH di bawah ini dengan path folder kamu di Google Drive
input_folder = '/content/drive/MyDrive/Tugas Akhir/Dataset/V2/Disney/'
output_folder = '/content/drive/MyDrive/Tugas Akhir/Dataset/V3/Disney/'

proses_folder_csv(input_folder, output_folder)
```
